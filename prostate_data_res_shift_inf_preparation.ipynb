```python
import h5py,numpy as np,matplotlib.pyplot as plt,glob,os
from PIL import Image
```

```python
pros_x_h5_path = '/srv/Data16tb/MRI_dataset/ProstateX_h5_dce15/'
```

```python
h5_files = sorted(glob.glob(pros_x_h5_path+'*.h5'))
print(len(h5_files))
```

```python
for i  in h5_files:
    patient_name = os.path.splitext(os.path.basename(i))[0]
    print(patient_name)

    with h5py.File(i,'r') as read_data:
        # print(read_data.keys())
        t2 = read_data['T2'][0,:,:,:]
        # print(t2.shape)
        adc = read_data['ADC'][0,:,:,:]
        for mid in range(7,11):
            mid_t2 = t2[:,:,mid]
            mid_adc = adc[:,:,mid]
            # print(mid_t2.shape)

                        # Normalize to 0-255 for saving as PNG
            norm_t2 = ((mid_t2 - np.min(mid_t2)) / (np.ptp(mid_t2)) * 255).astype(np.uint8)
            norm_adc = ((mid_adc - np.min(mid_adc)) / (np.ptp(mid_adc)) * 255).astype(np.uint8)
            # print("min max t2",np.min(norm_t2),np.max(norm_t2),norm_t2.shape)
            # print("min max adc",np.min(norm_adc),np.max(norm_adc),norm_adc.shape)
             
            # Save images
            t2_img = Image.fromarray(norm_t2)
            adc_img = Image.fromarray(norm_adc)
            print(f"{data_path_2d_adc}{patient_name}_ADC_slice{mid}.png")
            print(f"{data_path_2d_t2}{patient_name}_T2_slice{mid}.png")

            t2_img.save(f"{data_path_2d_t2}{patient_name}_T2_slice{mid}.png")
            adc_img.save(f"{data_path_2d_adc}{patient_name}_ADC_slice{mid}.png")
        # break        
```

```python
data_path_2d_adc = '/srv/Data16tb/Kishore/ResShift/testdata/user_own_data/Pros_datas/ADC_2d_prox_imgs/'
data_path_2d_t2 = '/srv/Data16tb/Kishore/ResShift/testdata/user_own_data/Pros_datas/T2_2d_prox_imgs/'
fol = [data_path_2d_adc,data_path_2d_t2]
for i in range(len(fol)):
    print(fol[i])
    
    if not os.path.exists(fol[i]):
        os.mkdir(fol[i])
```
